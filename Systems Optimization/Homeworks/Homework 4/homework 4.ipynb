```python
import gurobipy as gp
from gurobipy import *

# Create a new model
model = gp.Model("Sooner Inc Paints")

# Define decision variables
x1 = model.addVar(vtype=GRB.CONTINUOUS, lb=-GRB.INFINITY, ub=GRB.INFINITY, name="x1")
x2 = model.addVar(vtype=GRB.CONTINUOUS, lb=-GRB.INFINITY, ub=GRB.INFINITY, name="x2")
x3 = model.addVar(vtype=GRB.CONTINUOUS, lb=-GRB.INFINITY, ub=GRB.INFINITY, name="x3")
x4 = model.addVar(vtype=GRB.CONTINUOUS, lb=-GRB.INFINITY, ub=GRB.INFINITY, name="x4")

# Define the objective function
z = 25*x1 + 20*x2 + 2*x3 + 2*x4
model.setObjective(z)
model.modelSense = GRB.MINIMIZE
model.update()

# Add Constraints
model.addConstr(1*x1 + 2*x2 + 1*x3 >= 10)
model.addConstr(2*x1 + 2*x2 + 1*x4 >= 12)
model.addConstr(x1 >= 0)
model.addConstr(x2 >= 0)
model.addConstr(x3 >= 0)
model.addConstr(x4 <= 0)
model.update()

# Optimize the model
model.optimize()

# Print output
if model.status == GRB.OPTIMAL:
    print("Optimal value:", model.objVal, "USD")
    print("pigments: ", x1.X)
    print("water: ", x2.X)
    print("y3: ", x3.X)
    print("y4: ", x4.X)
```

```python
import gurobipy as gp
from gurobipy import *

# Create a new model for the dual problem
model = gp.Model("Sooner Inc Paints DUAL")

# Define decision variables for shadow prices
y1 = model.addVar(vtype=GRB.CONTINUOUS, lb=0, ub=GRB.INFINITY, name="y1")  # Dual for water constraint
y2 = model.addVar(vtype=GRB.CONTINUOUS, lb=0, ub=GRB.INFINITY, name="y2")  # Dual for pigment constraint

# Define the objective function (minimize resource cost)
z = 45*y1 + 40*y2
model.setObjective(z, GRB.MINIMIZE)
model.update()

# Add Constraints for the dual
# Coefficients come from the primal objective function for red, blue, and black paints
model.addConstr(1*y1 + 2*y2 >= 10)  # Red paint constraint
model.addConstr(2*y1 + 2*y2 >= 12)  # Blue paint constraint
model.addConstr((1/5)*y1 + (1/5)*y2 >= 9)  # Black paint constraint

# Optimize the model
model.optimize()

# Print output
if model.status == GRB.OPTIMAL:
    print("Optimal value (Cost):", model.objVal, "USD")
    print("Dual value for water: ", y1.X)
    print("Dual value for pigments: ", y2.X)
```

```python
import gurobipy as gp
from gurobipy import GRB

# Create a new model
model = gp.Model("Investment_Problem")

# Define decision variables (x_A and x_B for investments in stocks A and B)
x_A = model.addVar(vtype=GRB.CONTINUOUS, lb=0, name="x_A")
x_B = model.addVar(vtype=GRB.CONTINUOUS, lb=0, name="x_B")

# Define the objective function (minimize total investment)
model.setObjective(x_A + x_B, GRB.MINIMIZE)

# Add the constraints
# Profit constraint: 0.5 * x_A + 0.2 * x_B >= 20
model.addConstr(0.5 * x_A + 0.2 * x_B >= 20, "Profit")

# Investment ratio constraint: x_A <= 1.5 * x_B
model.addConstr(x_A <= 1.5 * x_B, "Investment Ratio")

# Minimum investment in stock B: x_B >= 5
model.addConstr(x_B >= 5, "Min Investment in B")

# Total investment constraint: x_A + x_B <= 100
model.addConstr(x_A + x_B <= 100, "Total Investment")

# Optimize the model
model.optimize()

# Print the optimal values of the decision variables
if model.status == GRB.OPTIMAL:
    print(f"Optimal investment in stock A: {x_A.X}")
    print(f"Optimal investment in stock B: {x_B.X}")
    print(f"Total investment: {x_A.X + x_B.X}")
```

```python
import gurobipy as gp
from gurobipy import GRB

# Create a new model for the dual problem
model = gp.Model("Dual_Problem")

# Define decision variables (y1, y2, y3, y4)
y1 = model.addVar(vtype=GRB.CONTINUOUS, lb=0, name="y1")
y2 = model.addVar(vtype=GRB.CONTINUOUS, ub=0, name="y2")
y3 = model.addVar(vtype=GRB.CONTINUOUS, lb=0, name="y3")
y4 = model.addVar(vtype=GRB.CONTINUOUS, ub=0, name="y4")

# Define the objective function (maximize)
model.setObjective(20*y1 + 5*y3 + 100*y4, GRB.MAXIMIZE)

# Add the constraints
model.addConstr(0.5*y1 + 1*y2 + 1*y4 <= 1, "c1")
model.addConstr(0.2*y1 - 1.5*y2 + 1*y3 + 1*y4 <= 1, "c2")

# Optimize the model
model.optimize()

# Print the optimal values of the dual variables
if model.status == GRB.OPTIMAL:
    print("Optimal value (Objective):", model.objVal)
    print(f"y1 = {y1.X}")
    print(f"y2 = {y2.X}")
    print(f"y3 = {y3.X}")
    print(f"y4 = {y4.X}")
```

```python
import gurobipy as gp
from gurobipy import GRB

# Create a new model
model = gp.Model("investment_portfolio")

# Decision variables: xA (investment in stock A), xB (investment in stock B)
xA = model.addVar(vtype=GRB.CONTINUOUS, name="xA", lb=0)
xB = model.addVar(vtype=GRB.CONTINUOUS, name="xB", lb=0)

# Set objective function: Minimize total investment xA + xB
model.setObjective(xA + xB, GRB.MINIMIZE)

# Add constraints
model.addConstr(0.5 * xA + 0.2 * xB >= 20, "ProfitConstraint")
model.addConstr(xA <= 1.5 * xB, "RatioConstraint")
model.addConstr(xB >= 5, "MinStockBConstraint")
model.addConstr(xA + xB <= 100, "BudgetConstraint")

# Optimize the model
model.optimize()

# Display the results
if model.status == GRB.OPTIMAL:
    print(f"Optimal Investment in Stock A: {xA.X}")
    print(f"Optimal Investment in Stock B: {xB.X}")
    print(f"Total Investment: {xA.X + xB.X}")
    
    # Sensitivity analysis for objective function coefficients
    print("\nSensitivity Analysis for Stock A and Stock B Objective Coefficients:")
    
    # Price range for stock A
    print(f"Stock A objective coefficient range: [{xA.SAObjLow}, {xA.SAObjUp}]")
    
    # Price range for stock B
    print(f"Stock B objective coefficient range: [{xB.SAObjLow}, {xB.SAObjUp}]")
    
    # Sensitivity analysis for RHS of total investment constraint
    budget_constraint = model.getConstrByName("BudgetConstraint")
    print(f"\nTotal Investment range: [{budget_constraint.SARHSLow}, {budget_constraint.SARHSUp}]")
else:
    print("No optimal solution found.")
```
