```python
import gurobipy as gp
import pandas as pd
from gurobipy import GRB
import numpy as np
import matplotlib.pyplot as plt
```

```python
scenarios = ['s1','s2']

demands = {
    's1': 50,   # demand for scenario 2
    's2': 100,  # demand for scenario 2
}

probabilities = {
    's1': 0.6,  # probability for scenario 1
    's2': 0.4,  # probability for scenario 2
}

c = 100

h = 5

q = 20
```

```python
# Create a new model
model = gp.Model("two_stage_optimization")

# Decision variables:
x = model.addVars(vtype=GRB.INTEGER, name="x")

# Decision variables
y = model.addVars(probabilities, vtype=GRB.INTEGER, name="y")

# Decision variables
z = model.addVars(probabilities, vtype=GRB.INTEGER, name="z")


# Objective function: Maximize total quality
total_cost = c*x + gp.quicksum(probabilities[i] h*y[i] + q* z(i)  for i in demands for j in probabilities)

# Add the budget constraint
model.addConstr(y[i] <= x, "capacity_constraint")

# Add the time constraint
model.addConstr(y[i] + z[i] <= demands[i], "time_constraint")
```
